```python
# 나의 코드
N = int(input())

if N == 1:
    print(1)
else:
    start = 2
    minDistance = 2
    section = 6

    while True:
        if start <= N <= start + section - 1:
            print(minDistance)
            break
        else:
            start += section
            minDistance += 1
            section += 6
```

```python
# 추천 코드
N = int(input())

minDistance = 1
range_end = 1

while N > range_end:
    range_end += minDistance * 6 
    minDistance += 1
print(minDistance)
```
