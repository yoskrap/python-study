```python
# 걷기: t초 | 달리기: t * (1 - k*0.01) 초
# 알고리즘 문제에서는 나눗셈,소수점을 안씀. 후에 100을 곱해서 정수로 비교해야함
```

```python
# 입력시간을 초 단위로 변환
def to_seconds(time):
    H, M, S = map(int, time.split(':'))
    return H*3600 + M*60 + S

current_time = input()
class_time = input()
t, k = map(int, input().split())

current_sec = to_seconds(current_time) # 현재 시간의 초 단위
class_sec = to_seconds(class_time) # 강의 시간의 초 단위

if current_sec > class_sec:
    print(0)
else:
    # 여유시간
    spare_time = class_sec - current_sec

    # 가는 시간
    # travel_time = t * (1 - k * 0.01) -- 이거 안씀
    travel_time = t * (100 - k)

    # 도착 가능 여부
    if spare_time * 100 >= travel_time:
        print(1) # 도착 가능
    else:
        print(0) # 지각
```
